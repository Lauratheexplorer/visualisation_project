```python
missing_values = ["n/a", "na", "--", ' ', '?', 'NaN']
data = pd.read_csv('mental_health.csv', na_values=missing_values)
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data
```

```python
data.isnull().sum()
```

```python
data.duplicated()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.to_csv('clean_mental_health.csv')
```

```python
male_data = data[data['Gender'].str.lower() == 'male']
female_data = data[data['Gender'].str.lower() == 'female']

# Randomly sample male respondents to match the number of female respondents
# Random state ensures reproducibility
balanced_male_data = male_data.sample(n=len(female_data), random_state=42)

# Combine the down sampled male data with the female data
balanced_data = pd.concat([balanced_male_data, female_data])

balanced_data.to_csv('balancedMH.csv', index=False)
```
