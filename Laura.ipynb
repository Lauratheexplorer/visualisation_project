```python
# import libraries and read the csv file
import numpy as np
import pandas as pd

data = pd.read_csv('balancedMH.csv')
data
```

```python
# describe the dataset
data.describe()
```

```python
# Convert timestamp column to datetime format and then create
# new columns with the date, month and day

from datetime import datetime as dt
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data['date'] = data['Timestamp'].dt.date
data['month'] = data['Timestamp'].dt.month
data['day'] = data['Timestamp'].dt.day
data
```

```python
# Create line graph showing total amount of records for each specific date
import matplotlib.pyplot as plt

daily_counts = data.groupby(data['date']).size()
daily_counts.head()

plt.figure(figsize=(12, 6))
plt.plot(daily_counts.index, daily_counts.values, marker='o', linestyle='-')
plt.title('Number of responses over time')
plt.xlabel('Date')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create line graph showing total amount of records for each month
monthly_counts = data.groupby(data['month']).size()
monthly_counts.head()

plt.figure(figsize=(12, 6))
plt.plot(monthly_counts.index, monthly_counts.values,
         marker='o', linestyle='-')
plt.title('Number of responses by month')
plt.xlabel('Month')
plt.ylabel('Number of Responses')
# Change labels from numbers to month names
plt.xticks([2, 4, 6, 8, 10, 12], ['February', 'April', 'June',
           'August', 'October', 'December'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create line graph showing total amount of records for each day of month
day_counts = data.groupby(data['day']).size()
day_counts.head()

plt.figure(figsize=(12, 6))
plt.plot(day_counts.index, day_counts.values, marker='o', linestyle='-')
plt.title('Number of responses by day')
plt.xlabel('Day of Month')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Drop the unwanted columns
try:
    data.drop(columns=['care_options', 'Social_Weakness', 'Occupation',
                       'Growing_Stress', 'treatment', 'month', 'day',
                       'date', 'Timestamp'], inplace=True)
except:
    next
data
```

```python
# Create pie charts to visualise proportions of 6 variables
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

cols_to_visualize = ['family_history', 'Mental_Health_History', 'Mood_Swings',
                     'Days_Indoors', 'Changes_Habits', 'Coping_Struggles']
counts = []

for col in cols_to_visualize:
    counts.append(data[col].value_counts())

fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()

for i, (col, count) in enumerate(zip(cols_to_visualize, counts)):
    axs[i].pie(count, labels=count.index, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(col)

plt.show()
```

```python
# Check all the unique values for a history of mental health
np.unique(data['Mental_Health_History'])

# Visualise the total of each of the mental health categories, split by family history
plt.figure(figsize=(10, 7))
sns.countplot(x='Mental_Health_History', data=data,
              palette='inferno', hue='family_history')
plt.title('Distribution mental health history')
plt.xlabel('Mental Health History ')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create new column which is a combination of family history and mental health history
# with 6 defined labels
correlation_detector = [
    (data['family_history'] == 'Yes') & (
        data['Mental_Health_History'] == 'Yes'),
    (data['family_history'] == 'Yes') & (
        data['Mental_Health_History'] == 'No'),

    (data['family_history'] == 'No') & (
        data['Mental_Health_History'] == 'Yes'),
    (data['family_history'] == 'No') & (data['Mental_Health_History'] == 'No'),
    (data['family_history'] == 'No') & (
        data['Mental_Health_History'] == 'Maybe'),
    (data['family_history'] == 'Yes') & (
        data['Mental_Health_History'] == 'Maybe')
]
values = ['Both yes', 'Family yes. Personal no.', 'Family no. Personal yes',
          'Both no', 'Family no. Personal maybe.', 'Family yes. Personal maybe.']
data['family_history_vs_personal'] = np.select(correlation_detector, values)
data.head()
```

```python
# Display the above information in a bar chart by counting the total amounts of each label
total_combined = data['family_history_vs_personal'].value_counts(
    normalize=False)
print('Total combined', total_combined)
x = ['Both no.', 'Family no. Personal maybe.', 'Personal yes. Family no.',
     'Family yes. Personal maybe.', 'Family yes. Personal no.', 'Both yes.']
fig, ax = plt.subplots()
ax.barh(x, total_combined, color='magenta')
plt.xlabel('total number')
plt.ylabel('Combined answers')
plt.title('Combination of personal history of mental health vs family history')
plt.show()
```

```python
# Check all the unique values for the new column
np.unique(data['family_history_vs_personal'])

# Visualise the amount for each of the categories in the new column, split by Gender
plt.figure(figsize=(10, 7))
sns.countplot(x='family_history_vs_personal', data=data,
              palette='inferno', hue='Gender')

plt.title(
    'History of either Personal or Family-related Mental Health issues by Gender.')
plt.xticks(rotation=45)
plt.xlabel('Family vs Personal')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Cross-tabulation
history_vs_time_indoors_ct = pd.crosstab(
    data['Days_Indoors'], data['family_history_vs_personal'])

# Convert counts to proportions for better comparison
history_vs_time_indoors_prop = history_vs_time_indoors_ct.div(
    history_vs_time_indoors_ct.sum(1), axis=0)
print(history_vs_time_indoors_prop)

history_vs_time_indoors_prop.plot(kind='bar', stacked=False, figsize=(10, 6))
plt.title('Family History vs Personal by Days Indoors')
plt.xlabel('Days Indoors')
plt.ylabel('Proportion with Family or Personal History')
plt.legend(title='Family History or Personal')
plt.xticks(rotation=45)
plt.ylim(top=0.3)
plt.tight_layout()
plt.show()
```
